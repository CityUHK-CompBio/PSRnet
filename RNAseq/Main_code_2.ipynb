```python
# fastqc
print('fastqc -f fastq -t %s -o 01_fastqc 01_rawdata/*.fastq.gz'%(thread))

print('echo \'************************************** fastqc finished! **************************************\'')
```

```python
# qc trim

with open('02_trim/02_trim.sh', 'w') as f:
    for i in sample:
        print('mkdir 02_trim/%s'%i)
        f.write('mkdir 02_trim/%s\n'%i)
        print('trim_galore -q 20 --phred33 --length 36 --fastqc -o 02_trim/%s --paired \
-a AGATCGGAAGAGCACACGTCTGAACTCCAGTCA -a2 AGATCGGAAGAGCGTCGTGTAGGGAAAGAGTG \
01_rawdata/%s*R1* 01_rawdata/%s*R2*'%(i,i,i))
        f.write('trim_galore -q 20 --phred33 --length 36 --fastqc -o 02_trim/%s --paired \
-a AGATCGGAAGAGCACACGTCTGAACTCCAGTCA -a2 AGATCGGAAGAGCGTCGTGTAGGGAAAGAGTG \
01_rawdata/%s*R1* 01_rawdata/%s*R2*\n'%(i,i,i))
    print('echo \'************************************** trim finished! **************************************\'')
    f.write('echo \'************************************** trim finished! **************************************\'')
```

```python
# mapping by STAR    ref use genebank

#STAR --runMode genomeGenerate --runThreadN 20 --sjdbGTFfeatureExon CDS --genomeDir ~/data/star_index/PS_1448A_gb_CP000058.1_STAR2.6 --genomeFastaFiles ~/data/star_index/PS_1448A_gb_CP000058.1_STAR2.6/CP000058.1.fasta --sjdbGTFfile ~/data/star_index/PS_1448A_gb_CP000058.1_STAR2.6/CP000058.1.gtf --sjdbOverhang 149 --genomeSAindexNbases 10  

#ulimit -n 65535

with open('03_star/03_star.sh', 'w') as f:
    for i in sample:
        print('mkdir 03_star/%s'%i)
        f.write('mkdir 03_star/%s\n'%i)
        print('STAR --runThreadN %s --genomeDir ~/data/star_index/PS_1448A_gb_CP000058.1_STAR2.6 \
--readFilesIn 02_trim/%s/%s_R1_val_1.fq.gz 02_trim/%s/%s_R2_val_2.fq.gz \
--outFileNamePrefix 03_star/%s/%s_ --outFilterType BySJout --outFilterMultimapNmax 1 --outFilterMismatchNmax 2 \
--outSAMtype BAM Unsorted SortedByCoordinate --quantMode TranscriptomeSAM GeneCounts --readFilesCommand zcat'%(thread,i,i,i,i,i,i))
        f.write('STAR --runThreadN %s --genomeDir ~/data/star_index/PS_1448A_gb_CP000058.1_STAR2.6 \
--readFilesIn 02_trim/%s/%s_R1_val_1.fq.gz 02_trim/%s/%s_R2_val_2.fq.gz \
--outFileNamePrefix 03_star/%s/%s_ --outFilterType BySJout --outFilterMultimapNmax 1 --outFilterMismatchNmax 2 \
--outSAMtype BAM Unsorted SortedByCoordinate --quantMode TranscriptomeSAM GeneCounts --readFilesCommand zcat\n'%(thread,i,i,i,i,i,i))
        print('samtools index 03_star/%s/%s_Aligned.sortedByCoord.out.bam'%(i,i))
        f.write('samtools index 03_star/%s/%s_Aligned.sortedByCoord.out.bam\n'%(i,i))

    print('echo \'************************************** mapping finished! **************************************\'')
    f.write('echo \'************************************** mapping finished! **************************************\'')
```

```python
# calculate tpm

# rsem-prepare-reference --gtf ~/data/rsem_star_index/PS_1448A_gb_CP000058.1/CP000058.1.gtf --star -p 20 ~/data/rsem_star_index/PS_1448A_gb_CP000058.1/CP000058.1.fasta ~/data/rsem_star_index/PS_1448A_gb_CP000058.1/PS_1448A

with open('03_tpm/03_tpm.sh', 'w') as f:
    for i in sample:
        print('mkdir 03_tpm/%s'%i)
        f.write('mkdir 03_tpm/%s\n'%i)
        print('rsem-calculate-expression --paired-end -p %s --bam 03_star/%s/%s_Aligned.toTranscriptome.out.bam \
~/data/rsem_star_index/PS_1448A_gb_CP000058.1/PS_1448A 03_tpm/%s/%s_'%(thread,i,i,i,i))
        f.write('rsem-calculate-expression --paired-end -p %s --bam 03_star/%s/%s_Aligned.toTranscriptome.out.bam \
~/data/rsem_star_index/PS_1448A_gb_CP000058.1/PS_1448A 03_tpm/%s/%s_\n'%(thread,i,i,i,i))

    print('echo \'************************************** mapping finished! **************************************\'')
    f.write('echo \'************************************** mapping finished! **************************************\'')
```
