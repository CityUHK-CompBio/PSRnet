```python
# bedtools

with open('04_bedtools/04_bedtools.sh','w') as f:
    for i in gene:
        print('bedtools intersect -a 04_macs/%s_1/%s_1_peaks.narrowPeak -b 04_macs/%s_2/%s_2_peaks.narrowPeak \
-wo > 04_bedtools/%s.overlaps.bed'%(i,i,i,i,i))
        f.write('bedtools intersect -a 04_macs/%s_1/%s_1_peaks.narrowPeak -b 04_macs/%s_2/%s_2_peaks.narrowPeak \
-wo > 04_bedtools/%s.overlaps.bed\n'%(i,i,i,i,i))
        
    print('echo \'************************************** bedtools finished! **************************************\'')
    f.write('echo \'************************************** bedtools finished! **************************************\'')
```
